```python
# Gera um DataFrame com números de 0 a 100 milhões
data = pd.DataFrame({'price': range(0, 100000001)})
data.tail(1)
```

```python
%%time
# Utiliza o apply() para aplicar o desconto de 5%
data["apply"] = data["price"].apply(lambda x: x * 0.95)
data.tail(1)
```

```python
# Excluí a coluna "apply" para o próximo teste
data.drop(columns=["apply"], inplace=True)
```

```python
%%time
# Utiliza vetorização para aplicar o desconto de 5%
data["vectorized"] = data["price"] * 0.95
data.tail(1)
```
